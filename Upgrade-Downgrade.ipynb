```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
```

```python
url="http://finance.yahoo.com/q/ud?s=AAPL"
response=urllib.request.urlopen(url)
soup=BeautifulSoup(response,"html.parser")
```

```python
print(soup.prettify())
```

```python
AnalystDec=[]
AnalystDec=soup.find_all("td", "yfnc_tabledata1")
```
